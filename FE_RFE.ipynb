```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'].isnull().sum()
```

```python
# Deleting NaN values
df.dropna(how='any', inplace=True)
```

```python
df.info()
```

```python
df.drop(['customerID'], axis='columns', inplace=True)
```

```python
# Dummy Encoding
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
df.Churn_Yes.value_counts()/len(df)*100
```

```python
X = df.drop('Churn_Yes', axis=1)
y = df['Churn_Yes']
```

```python
y
```

```python
# Split the data into training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### 5 features

```python
# Apply Recursive Feature Elimination (RFE) to select 5 features
lr = LogisticRegression()
rfe = RFE(lr, n_features_to_select=5)
X_train_rfe = rfe.fit(X_train, y_train)
```

```python
# Get the selected features

selected_features = X_train.columns[X_train_rfe.support_]
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]
```

```python
# Build the model

model_orig = LogisticRegression()
model_orig.fit(X_train, y_train)
y_pred_orig = model_orig.predict(X_test)

accuracy_orig = accuracy_score(y_test, y_pred_orig)
print("Accuracy of original model:", round(accuracy_orig*100, 2))
print("Report of Original model:", classification_report(y_test, y_pred_orig))
```

```python
# Build the RFE based model on 5 predictors

model_rfe = LogisticRegression()
model_rfe.fit(X_train_selected, y_train)
y_pred_rfe = model_rfe.predict(X_test_selected)

accuracy_rfe = accuracy_score(y_test, y_pred_rfe)
print("Accuracy of RFE model:", round(accuracy_rfe*100, 2))
print("Report of RFE model:", classification_report(y_test, y_pred_rfe))
```
