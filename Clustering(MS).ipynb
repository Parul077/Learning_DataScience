```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift
```

```python
dt = pd.read_csv('/content/mall.csv')
```

```python
dt.head()
```

```python
X = dt.iloc[:,[3,4]].values
```

```python
# Fitting the mean shift clustering to the mall dataset

ms = MeanShift(bandwidth=20)

# Adjust the bandwidth parameter to influence the number of clusters

ms.fit(X)
labels = ms.labels_
cluster_centers = ms.cluster_centers_

```

```python
# visualizing the clusters

n_clusters = len(np.unique(labels))
colors = ['red', 'blue', 'green', 'yellow', 'cyan', 'purple', 'pink']

for i in range(n_clusters):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], s=100, c=colors[i], label=f'Cluster {i+1}')

plt.title('Clusters of Clients')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score (0-100)')
plt.legend()
plt.show()
```
